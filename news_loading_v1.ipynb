```python
def get_news_text(url):
    try:
        response = urllib.request.urlopen(url)
    except HTTPError:
        return ''
    else:
        html = response.read().decode('utf-8')
        soup = BeautifulSoup(html, features="lxml")
        doc = []
        content = ''
        p_list = soup.find_all('p')
        p_list_new = []
        for i in p_list:
            try:
                if i['class'] == ['css-g5piaz', 'evys1bk0']:
                    p_list_new.append(i)
            except KeyError:
                continue
        for p in p_list_new:
            for text in p.strings:
                doc.append(text)
        content = ''.join(doc)
        return content
```

```python
def news_get_request(url):
    results = requests.get(url+api_key)
    json_data = results.json()
    return json_data
```

```python
def get_most_viewed_30():
    json_data = news_get_request(viewed_url)
    news_json = {'data' : []}
    for news in json_data['results']:
        data = {}
        data['id'] = news['id']
        data['url'] = news['url']
        data['publish_data'] = news['published_date']
        data['section'] = news['section']
        data['title'] = news['title']
        data['abstract'] = news['abstract']
        data['keywords'] = news['adx_keywords']
        data['author'] = news['byline']
        data['text'] = get_news_text(news['url'])
        news_json['data'].append(data)
    return news_json
```

```python
def get_initial_news(year, month):
    start = time.time()
    request_url = 'https://api.nytimes.com/svc/archive/v1/{year}/{month}.json?api-key='.format(year = year, month = month)
    json_data = news_get_request(request_url)
    news_json = {'data' : []}
    raw_data = json_data['response']['docs']
    total = len(raw_data)
    i = 1
    for news in raw_data:
        start_round = time.time()
        data = {}
        # get author
        author = news['byline']['person']
        if author == []:
            continue
        else:
            name_comp = list(author[0].keys())[:3]
            author_list = []
            for person in author:
                name_list = []
                for comp in name_comp:
                    if person[comp] != None:
                        name_list.append(person[comp])
                name = ' '.join(name_list)
                author_list.append(name)
        data['author'] = author_list
        # get caregory
        section = news['section_name']
        data['section'] = section
        # get abstract
        data['abstract'] = news['abstract'] 
        # get url
        data['url'] = url
        # get tital
        data['title'] = news['headline']['main']
        # get keywords
        kw_lsit = []
        for kw in news['keywords']:
            kw_lsit.append(kw['value'])
        data['keywords'] = kw_lsit
        # get publish date
        date = news['pub_date']
        timestamp = date.split('T')
        data['publish_data'] = timestamp
        # get id
        raw_id = news['_id']
        real_id = raw_id.split('/')[-1]
        data['id'] = real_id
        news_json['data'].append(data)
        end_round = time.time()
        print('completing------{}/{}, using time: {}'.format(i, total, end_round - start_round))
        i += 1
    end = time.time()
    print('total # of news in dic：{}. total time: {}'.format(len(news_json['data']), end - start))
    return news_json
```

```python
def clean_empty_text(data):
    data_raw = data['data']
    data_new = data_raw[:]
    count = 0
    for i in range(len(data_raw)):
        cur = data_raw[i]
        if 'text' not in cur.keys():
            data_new.pop(i-count)
            count += 1
    return data_new
```

```python
data_new = clean_empty_text(data)
```

```python
def save_dic_to_json(data, year, month):
    json_data = json.dumps(data)
    json_file = open('{}-{}-news.json'.format(year, month), 'w')
    json_file.write(json_data)
    json_file.close()
    return True
```
